```python
col_to_drop =['0-9 y/o % of total pop', '10-19 y/o % of total pop', '20-29 y/o % of total pop', '30-39 y/o % of total pop', '40-49 y/o % of total pop', '50-59 y/o % of total pop', '60-69 y/o % of total pop', '70-79 y/o % of total pop', '80+ y/o % of total pop']
raw_data = raw_data.drop(columns=col_to_drop)
raw_data
```

```python
variables = ['Urban_rural_code', 'STATE_NAME']

fig, axis = plt.subplots(figsize=(35, 15), nrows=2)
axis = axis.flatten()

for i, variable in enumerate(variables):
    sns.countplot(ax=axis[i], data=raw_data, x=variable, hue=variable)


plt.tight_layout()
plt.show()
```

```python
variables = ['anycondition_number', 'Obesity_number', 'Heart disease_number', 'COPD_number', 'diabetes_number', 'CKD_number']

for variable in variables:
    sns.catplot(data=raw_data, x='Urban_rural_code', y=variable, hue='Urban_rural_code')
    sns.catplot(data=raw_data, x='Urban_rural_code', y=variable, hue='Urban_rural_code', kind='box')
```
