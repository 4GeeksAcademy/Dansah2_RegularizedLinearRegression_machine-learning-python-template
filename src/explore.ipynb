```python
X = raw_data.drop('anycondition_number', axis=1)
y = raw_data['anycondition_number']
```

```python
# use either feature ranking / recursive feature elimination or SelectKBest to select the best features.
def choose_features(X, y, num_features):
    run_RFE = False

    if run_RFE:
        # obtain the feature names
        feature_names = [f'Feature_{i}' for i in range(X.shape[1])]
        estimator = SVR(kernel='linear')
        selector = RFE(estimator, step=1)

        selector.fit(X, y)

        selected_features = X.columns[selector.support_]

        print("Selected Features:", selected_features.tolist())

        return selected_features
    else:
        X_copy = X.copy()
        X_copy = X_copy - X_copy.min() + 1
    
        X_scaled = pd.DataFrame(power_transform(X_copy, method='box-cox'), columns=X.columns)
        
        selector = SelectKBest(f_regression, k=num_features)
        X_k_best = selector.fit_transform(X_scaled, y)
        
        feature_names = X.columns[selector.get_support()]
        X_k_best = pd.DataFrame(X_k_best, columns=feature_names)

        print(X_k_best.columns.tolist())
        
        return X_k_best.columns.tolist()
        
```

```python
K_best_features = choose_features(X, y, 95)
```

```python
X = raw_data[K_best_features]

X.head()
```

```python
def get_outliers(df, feature, is_categorical=False):
    if is_categorical:
        pass
    else:
        stats = df[feature].describe()
        iqr = stats['75%'] - stats['25%']
        upper_limit = stats['75%'] + 1.5 * iqr
        lower_limit = stats['25%'] - 1.5 * iqr
        up_out = df[df[feature] > upper_limit]
        low_out = df[df[feature] < lower_limit]
        print(f"{feature} Upper limit: {round(upper_limit, 2)}, Lower limit: {round(lower_limit, 2)}")

        return up_out, low_out
```

```python
def choose_scaler(training_data, testing_data, type_scale):

    if type_scale == 'b':
        training_data = training_data - training_data.min() + 1
        testing_data = testing_data - testing_data.min() + 1
        
        scaled_training = pd.DataFrame(power_transform(training_data, method='box-cox'), columns=training_data.columns)
        scaled_testing = pd.DataFrame(power_transform(testing_data, method='box-cox'), columns=testing_data.columns)

    if type_scale == 'mm':
        scaler = MinMaxScaler()

        scaled_training = scaler.fit_transform(training_data)
        scaled_testing = scaler.fit_transform(testing_data)

    return scaled_training, scaled_testing
```

```python
scaled_X_train, scaled_X_test = choose_scaler(X_train, X_test, 'b')
```

```python
scaled_X_train
```

```python
lasso_model = Lasso()
lasso_model.fit(scaled_X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = lasso_model.predict(scaled_X_test)

print(r2_score(y_pred, y_test))
```

```python
ridge_model = Ridge()
ridge_model.fit(scaled_X_train, y_train)
```

```python
y_pred = ridge_model.predict(scaled_X_test)

print(r2_score(y_pred, y_test))
```

## Tune the lasso model using grid search

```python
# create gridsearch for Lasso
hyperparams = {
    'alpha': [0.001, 0.01, 0.1, 1.0, 10, 100, 1000],
    'selection': ['cyclic', 'random']
}

grid = GridSearchCV(lasso_model, hyperparams, scoring='r2', cv=5)
grid
```

```python
#create the grid search
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(scaled_X_train, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```
